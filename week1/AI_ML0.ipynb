# 3. Python Data Science Toolbox

```python
def tuple_ex():
    t = (1,2,3)
    return t
a,b,c = tuple_ex()
print(a,b,c)
```

```python
import builtins
dir(builtins)
```

```python
#flexible arguments : *args

def f(*args):
    for i in args:
        print(i)
        
f(1,2,3,4)

#flexible arguments : **kwargs i.e. dict 

def g(**kwargs):
    for key, value in kwargs.items():
        print(key, " ", value)
g(country = 'spain', capital = 'madrid', pop = 123456)
```

```python
#anonymous function : lambda
#map(func, seq) -> applies a function to all items in a list 
number_list = [1,2,3]
y = map(lambda x: x**2, number_list)
print(list(y))
```

```python
#iterators 
#iterable: list, string, dict
name = "ronaldo"
it = iter(name)
print(next(it)) #next iteration 
print(*it) #remaining iteration
```

```python
#zip
list1 = [1,2,3,4]
list2 = [5,6,7,8]

z = zip(list1, list2) #makes list of tuples 
print(z) #zip object
z = list(z)
print(z) #list made by zip operator

unzip = zip(*z)
unlist1, unlist2 = list(unzip) #unzip returns tuple 
print(unlist1, unlist2) 
```

```python
#list comprehension

num1 = [1,2,3]
num2 = [ 2 * i for i in num1]
print(num2)

#using conditionals 
num2 = [ i**2 if i == 2 else i - 5 if i < 3 else i+5 for i in num1]
print(num2)
```

### Resampling Pandas Time Series 

* resampling : statistical method over different time intervals
    - "M" = month 
    - "A" = year
* Downsampling: reduce date time rows to slower (daily -> monthly)
* Upsampling: increase date time rows to faster (daily -> hourly)
* interpolate : insert values according to method; linear, time, index ... 

```python
data2.resample('A').mean()
```

```python
data2.resample('M').mean()
```

```python
data2.resample('M').mean().interpolate('linear')
```

### Pivoting Data Frame 
* pivoting : reshape tool

```python
dic = {"treatment":["A","A","B","B"],"gender":["F","M","F","M"],"response":[10,45,5,9],"age":[15,4,72,65]}
df = pd.DataFrame(dic)
df
```

```python
# pivoting 
df.pivot(index = 'treatment', columns = 'gender', values = 'response')
pd.melt(df, id_vars = 'treatment', value_vars = ['age', 'response'])
```
